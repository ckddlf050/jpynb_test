```python
import numpy as np
x = np.arange(4)
y = [100,400,900,1200]

plt.bar(x,y,color = ['r','g','b'],width = 0.7)
plt.plot(x,y)

plt.xticks(x,['2018','2019','2020','20221'])

plt.show()
```

```python
import numpy as np
y = np.arange(3)
x = [100,400,900]

plt.barh(y,x,color = ['r','g','b'],height=0.7)
plt.plot(x,y)

plt.yticks(y,['2018','2019','2020'])

plt.show()
```

```python
n = 50
x = np.random.rand(n)
y = np.random.rand(n)

area = (30 * np.random.rand(n)) **2
colors = np.random.random(n)

plt.scatter(x,y,area,colors,cmap='Spectral',alpha=0.5)
plt.show()
```

```python
ratio =[34,32,16,18]
labels=['Apple','Banana','Melon','Grapes']
wedgepros = {'width' : 0.7, 'edgecolor':'k','linewidth':1}

plt.pie(ratio,labels=labels,autopct='%.1f%%',explode=[0,0.1,0,0.1],colors=['r','g','b','y'],wedgeprops=wedgepros,startangle=90)
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales_2019 = [100,120,140,110,130,150,160,170,180,200,190,210]
sales_2020 = [90,110,130,120,140,160,170,160,150,180,200,190]

plt.plot(months,sales_2019,label = "2019")
plt.plot(months,sales_2020,label = "2020")

plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales Comparison(2019-2020)')

plt.legend(loc='best')

plt.show()
```

```python
x = np.arange(5)

categories = ['Category 1','Category 2','Category 3','Category 4','Category 5']
data = [20, 35, 15, 27, 45]


plt.bar(x,data, width = 0.8)
plt.xticks(x,categories)
plt.yticks(np.arange(0,51,10))
plt.grid(True)
plt.title('Bar Cahart')
plt.xlabel('Value')
plt.ylabel('Categories',rotation=45)
plt.show()
```

```python
labels = ['Apple','Banan','Mango','Blueberry']
sizes=[15,30,45,10]
explode=(0,0.1,0,0)

plt.pie(sizes,labels=labels,autopct='%.lf%%',explode=explode,colors=['b','orange','g','r'],startangle=90,shadow=True)
```

```python
x = np.linspace(0, 10, 100)

# sin cos   
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(y_sin,color = 'b',label = 'sin(x)')
plt.plot(y_cos,color = 'r',label= 'cos(x)')
plt.grid()

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Sine and cosine Function')
plt.legend()

plt.show()
```
