```python
import requests
from bs4 import BeautifulSoup

search_query = input("검색어를 입력하세요>>>")
response = requests.get(f"https://search.naver.com/search.naver?where=news&sm=tab_jum&query={search_query}")
html = response.text
soup = BeautifulSoup(html, 'html.parser')

# 뉴스 제목과 링크 추출
links = soup.select(".news_tit")
print(f"검색어: {search_query}")
for link in links:
    title = link.text
    url = link.attrs['href']
    print(title, url)
```

```python
import requests
from bs4 import BeautifulSoup
import openpyxl
```

```python
response = requests.get('https://finance.naver.com/marketindex/')
html = response.text
soup = BeautifulSoup(html, "html.parser")

ul = soup.find('ul', class_="data_lst", id="exchangeList")


for li in ul.find_all('li'):
    
    country_full = li.find('span', class_='blind').text
    country = country_full.split()[-1]  

    price = li.find('span', class_='value').text


    print(f"{country}: {price}")
```

```python
data = []
data.append([country,price])

bb=openpyxl.Workbook()

bs = bb.create_sheet('test')
bs.append (['통화','환율'])

for i in data:
    bs.append(i)

bb.save('test_cp.xlsx')
```
