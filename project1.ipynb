```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from datetime import datetime
import time
import csv

# options 객체를 만들지 않으면 실행시 브라우저 창이 꺼짐
options = Options()
options.add_argument("--start-maximized")  # 창크기 최대화
options.add_experimental_option('detach', True)

driver = webdriver.Chrome(options=options)  # 웹 드라이버 실행

# ABC News 검색 URL
url = "https://abcnews.go.com/search?searchtext=AI&after=year"
driver.get(url)
time.sleep(5)
```

```python
# 최대 페이지 수
max_page = 89


with open("abc_news.csv", "w", newline="", encoding="utf-8-sig") as file:
    writer = csv.writer(file)
    writer.writerow(["제목", "날짜", "요약"]) 

    for page in range(1, max_page + 1):
        try:
            
            url = f"https://abcnews.go.com/search?searchtext=AI&after=year&page={page}"
            driver.get(url)
            time.sleep(2)  

            # 기사 크롤링
            articles = driver.find_elements(By.CSS_SELECTOR, ".ContentRoll__Item")
            if not articles:
                print(f"{page}페이지에 기사가 없습니다.")
                continue

            for article in articles:
                try:
                    # 제목
                    title_element = article.find_element(By.CSS_SELECTOR, "a")
                    title = title_element.text.strip()
                    link = title_element.get_attribute("href")

                    summary = "요약 없음"  # 기본값 설정
                    try:
                        summary_element = article.find_element(By.CSS_SELECTOR, ".ContentRoll__Desc")
                        summary = summary_element.text.strip()
                    except:
                        print("요약 내용이 없습니다.")  # 요약 내용이 없는 경우 예외 처리

                    # 날짜
                    date_element = article.find_elements(By.CSS_SELECTOR, ".ContentRoll__Date.ContentRoll__Date--tabletPlus")
                    if date_element:
                        r_date = date_element[0].text
                        try:
                            
                            p_date = datetime.strptime(r_date, "%B %d, %Y")
                            date = p_date.strftime("%Y-%m-%d")
                        except ValueError:
                           
                            date = "날짜 없음"
                    else:
                        # 날짜 정보가 없는 경우 처리
                        date = "날짜 없음"

                    # CSV에 데이터 기록
                    writer.writerow([title, date, summary])

                    # 출력
                    print(f"제목: {title}")
                    print(f"날짜: {date}")
                    print(f"요약: {summary}\n")

                except Exception as e:
                    print("Error occurred in article parsing:", e)

        except Exception as e:
            print(f"Error occurred on page {page}: {e}")
```
