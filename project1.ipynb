```python
# 최대 페이지 수//중복 제거 안함//따로 처리 해야함
max_page = 73


with open("abc_news_test.csv", "w", newline="", encoding="utf-8-sig") as file:
    writer = csv.writer(file)
    writer.writerow(["제목", "날짜", "요약"]) 

    for page in range(1, max_page + 1):
        try:
            
            url = f"https://abcnews.go.com/search?searchtext=Artificial%2520Intelligence&after=year&sort=date&page={page}"
            driver.get(url)
            time.sleep(2)

            # 기사 크롤링
            articles = driver.find_elements(By.CSS_SELECTOR, ".ContentRoll__Item")
            if not articles:
                
                continue

            for article in articles:
                try:
                    # 제목
                    title_element = article.find_element(By.CSS_SELECTOR, "a")
                    title = title_element.text.strip()
                    link = title_element.get_attribute("href")

                    summary = "요약 없음"  # 기본값 설정
                    try:
                        summary_element = article.find_element(By.CSS_SELECTOR, ".ContentRoll__Desc")
                        summary = summary_element.text.strip()
                    except:
                        print("요약 내용이 없습니다.")  # 요약 내용이 없는 경우 예외 처리

                    # 날짜
                    date_element = article.find_elements(By.CSS_SELECTOR, ".ContentRoll__Date.ContentRoll__Date--tabletPlus")
                    if date_element:
                        r_date = date_element[0].text
                        try:
                            
                            p_date = datetime.strptime(r_date, "%B %d, %Y")
                            date = p_date.strftime("%Y-%m-%d")
                        except ValueError:
                           
                            date = "날짜 없음"
                    else:
                        # 날짜 정보가 없는 경우 처리
                        date = "날짜 없음"

                    # CSV에 데이터 기록
                    writer.writerow([title, date, summary])

                    # 출력
                    print(f"제목: {title}")
                    print(f"날짜: {date}")
                    print(f"요약: {summary}\n")

                except Exception as e:
                    print("Error occurred in article parsing:", e)

        except Exception as e:
            print(f"Error occurred on page {page}: {e}")
```

```python
max_page = 73

with open("abc_news_links.csv", "w", newline="", encoding="utf-8-sig") as link_file:
    link_writer = csv.writer(link_file)
    link_writer.writerow(["제목", "날짜", "요약", "링크"])
    
    for page in range(1, max_page + 1):
        try:
            url = f"https://abcnews.go.com/search?searchtext=Artificial%2520Intelligence&after=year&sort=date&page={page}"
            driver.get(url)
            time.sleep(2)

            articles = driver.find_elements(By.CSS_SELECTOR, ".ContentRoll__Item")
            if not articles:
                continue

            for article in articles:
                try:
                    title_element = article.find_element(By.CSS_SELECTOR, "a")
                    title = title_element.text.strip()
                    link = title_element.get_attribute("href")

                    summary = "요약 없음"
                    try:
                        summary_element = article.find_element(By.CSS_SELECTOR, ".ContentRoll__Desc")
                        summary = summary_element.text.strip()
                    except:
                        print("요약 내용이 없습니다.")

                    date_element = article.find_elements(By.CSS_SELECTOR, ".ContentRoll__Date.ContentRoll__Date--tabletPlus")
                    if date_element:
                        r_date = date_element[0].text
                        try:
                            p_date = datetime.strptime(r_date, "%B %d, %Y")
                            date = p_date.strftime("%Y-%m-%d")
                        except ValueError:
                            date = "날짜 없음"
                    else:
                        date = "날짜 없음"

                    link_writer.writerow([title, date, summary, link])
                    print(f"링크 저장 완료: {link}")

                except Exception as e:
                    print("Error occurred in link parsing:", e)

        except Exception as e:
            print(f"Error occurred on page {page}: {e}")
```

```python
import time
import csv
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# 옵션 설정
options = Options()
options.add_argument("--start-maximized")
options.add_experimental_option('detach', True)

# 웹 드라이버 실행
driver = webdriver.Chrome(options=options)
driver.set_page_load_timeout(30)  # 페이지 로드 타임아웃 설정


max_retries = 3
wait_time = 120  # 타임아웃 시간

# 실패한 URL 기록용 파일
with open("failed_links.csv", "w", newline="", encoding="utf-8-sig") as error_file:
    error_writer = csv.writer(error_file)
    error_writer.writerow(["제목", "링크", "오류 메시지"])

    # 뉴스 링크 파일 읽기
    with open("abc_news_links.csv", "r", encoding="utf-8-sig") as link_file:
        reader = csv.DictReader(link_file)
        
        # 본문 크롤링 결과 저장
        with open("abc_news_contents.csv", "w", newline="", encoding="utf-8-sig") as content_file:
            writer = csv.writer(content_file)
            writer.writerow(["제목", "날짜", "요약", "링크", "본문"])

            for row in reader:
                retries = 0
                content_fetched = False

                while retries < max_retries:
                    try:
                        driver.get(row["링크"])
                        time.sleep(2)

                        # 본문 CSS Selector로 내용 가져오기
                        content_element = driver.find_element(By.CSS_SELECTOR, ".xvlfx.ZRifP.TKoO.eaKKC.EcdEg.bOdfO.qXhdi.NFNeu.UyHES")
                        content = content_element.text.strip()
                        
                        writer.writerow([row["제목"], row["날짜"], row["요약"], row["링크"], content])
                        print(f"Content fetched for {row['제목']}")
                        content_fetched = True
                        break  # 성공하면 루프 종료
                    except TimeoutException:
                        retries += 1
                        print(f"Retry {retries}/{max_retries} for {row['제목']}: Timeout occurred.")
                        time.sleep(10)  # 재시도 전 대기 시간 추가
                    except NoSuchElementException:
                        print(f"Content not found for {row['제목']}. Skipping...")
                        error_writer.writerow([row["제목"], row["링크"], "Content not found"])
                        break  # 다음 URL로 넘어가기
                    except Exception as e:
                        print(f"Unexpected error for {row['제목']}: {e}")
                        retries += 1
                        time.sleep(10)  # 재시도 전 대기 시간 추가

                if not content_fetched and retries == max_retries:
                    print(f"Failed to fetch content for {row['제목']} after {max_retries} retries.")
                    error_writer.writerow([row["제목"], row["링크"], "Timed out after 3 retries"])
                    continue  # 다음 URL로 이동
```

```python

#x

with open("abc_news_links.csv", "r", encoding="utf-8-sig") as link_file:
    reader = csv.DictReader(link_file)
    with open("abc_news_contents.csv", "w", newline="", encoding="utf-8-sig") as content_file:
        writer = csv.writer(content_file)
        writer.writerow(["제목", "날짜", "요약", "본문"])

        for row in reader:
            try:
                link = row["링크"]
                driver.get(link)
                time.sleep(2)  # 페이지 로딩 대기

                try:
                    # 본문 크롤링 (CSS 선택자는 사이트에 따라 다름)
                    content_element = driver.find_element(By.XPATH, '//p[contains(@class, "EkqkG")]')
                    content = content_element.text.strip()
                except:
                    content = "본문 없음"

                writer.writerow([row["제목"], row["날짜"], row["요약"], content])
                print(f"본문 크롤링 완료: {row['제목']}")

            except Exception as e:
                print(f"Error occurred while fetching content for {row['제목']}: {e}")
```
